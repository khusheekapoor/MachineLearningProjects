**Meta-Data of the IRIS Dataset**

The IRIS dataset is a benchmark dataset collected in 1988 for multiclass classification based on the following attributes:

- SepalLengthCm: length of the sepal in cm, continuous variable
- SepalWidthCm: width of the sepal in cm, continuous variable
- PetalLengthCm: length of the petal in cm, continuous variable
- PetalWidtchCm: width of the petal in cm, continuous variable
- Species: three species of iris: setosa, versicolor, virginica, categorical variable

The dataset has 50 instances of each species, so the classes are balanced.

### Q3. Use the IRIS data set to build a logistic regression model (using sklearn) and predict the class label ‘Species’. Divide the dataset into training and test set (70,30) using train_test_split method in sklearn.

To solve Question 3, we first split the data into independent variables (x) and independent variable (y).

```python
# splitting the data into independent and dependent variables
x = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = df['Species']
```

Next, we use the train_test_split() function from the sklearn library and divide the dataset into training and testing sets. We set the train_size parameter to 0.7 to ensure that 70% of the dataset is put into training.

```python
# diving the dataset into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=105)
```

After that, we use the LogisticRegression() module from the sklearn library to fit the training set on the model and predict the values of the testing set.

```python
from sklearn.linear_model import LogisticRegression

# building the model
logreg = LogisticRegression()
logreg.fit(x_train, y_train)

# predicting from testing set
y_pred = logreg.predict(x_test)
```

### Q4. Analysis and visualize the performance of the classifier using metrics, confusion matrix.

To solve Question 4, we first use the confusion_matrix() function from the sklearn library to obtain the confusion matrix. Next, we prepare the labels for the matrix to make it more readable. We then plot the confusion matrix using the heatmap() function from the Seaborn library.

```python
from sklearn.metrics import confusion_matrix

# obtaining the confusion matrix
cf_matrix = confusion_matrix(y_test, y_pred)

# preparing the labels
labels = ['Correct Classification', 'Misclassification', 'Misclassification', 
          'Misclassification', 'Correct Classification', 'Misclassification', 
          'Misclassification', 'Misclassification', 'Correct Classification']
values = ['{0:0.0f}'.format(value) for value in cf_matrix.flatten()]
annotations = [f'{v1}\n{v2}' for v1, v2 in zip(labels, values)]
annotations = np.asarray(annotations).reshape(3,3)
axxlabels = ['Predicted Setosa', 'Predicted Versicolor', 'Predicted Virginica']
axylabels = ['Actual Setosa', 'Actual Versicolor', 'Actual Virginica']

# plotting the heatmap
plt.figure(figsize=(10,7))
sns.heatmap(cf_matrix, annot=annotations, fmt='', xticklabels=axxlabels, yticklabels=axylabels)
```

As we can see, there is only 1 misclassified case, wherein the model has classified versicolor as virginica. This means that the model is very good.

To further evaluate the model, we will use the score() function to obtain the accuracy of the model on the testing set.

```python
# obtaining the accuracy of the model
print(str.format('Accuracy: {:.2f}%', logreg.score(x_test, y_test)*100))
```

As we can see, the accuracy of the model is 97.78%, which is very good.

### Q5/ 6. Use the IRIS data and KNeighborsClassifier (using sklearn) and predict the class label ‘Species’ for k value between 2 and 20. Divide the dataset into training and test set (70,30) using train_test_split method in sklearn. Identify the best k (for k between 2 and 20) for the model built.

To solve Questions 5 and 6, we use the KNeighborsClassifier() module from the sklearn library and iterate over the value of k from 2 to 20 and pass it to the n_neighbors parameter. Then we compute the test error using all the instances where the predicted class is not equal to the actual class of all the values of k. After that, we use the lineplot() function from the Seaborn library to plot the variation of error with the value of k.

```python
from sklearn.neighbors import KNeighborsClassifier
error = []

# iterating over the value of k and building models
for k in range(2,20):    
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    pred = knn.predict(x_test)
    error.append(np.mean(pred!=y_test))
    
# plotting the variation of error with the value of k
plt.figure()
sns.lineplot(x=range(2,20), y=error)
plt.title('Value of K vs Error')
plt.xlabel('Value of K')
plt.xlim(2,20)
plt.ylabel('Error')
```

As we can see, the error is 0 for values between 2 to 8, and then fluctuates afterwards. Since the size of the dataset is small, we choose the value of k to be 5 so as to neither overfit nor underfit.

After that, we use the KNeighborsClassifier() module from the sklearn library to fit the training set on the model and predict the values of the testing set by passing the value of k to be 5.

```python
# building the model
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)

# predicting from testing set
y_pred = knn.predict(x_test)
```

To evaluate the model, we first use the confusion_matrix() function from the sklearn library to obtain the confusion matrix. Next, we prepare the labels for the matrix to make it more readable. We then plot the confusion matrix using the heatmap() function from the Seaborn library.

```python
from sklearn.metrics import confusion_matrix

# obtaining the confusion matrix
cf_matrix = confusion_matrix(y_test, y_pred)

# preparing the labels
labels = ['Correct Classification', 'Misclassification', 'Misclassification', 
          'Misclassification', 'Correct Classification', 'Misclassification', 
          'Misclassification', 'Misclassification', 'Correct Classification']
values = ['{0:0.0f}'.format(value) for value in cf_matrix.flatten()]
annotations = [f'{v1}\n{v2}' for v1, v2 in zip(labels, values)]
annotations = np.asarray(annotations).reshape(3,3)
axxlabels = ['Predicted Setosa', 'Predicted Versicolor', 'Predicted Virginica']
axylabels = ['Actual Setosa', 'Actual Versicolor', 'Actual Virginica']

# plotting the heatmap
plt.figure(figsize=(10,7))
sns.heatmap(cf_matrix, annot=annotations, fmt='', xticklabels=axxlabels, yticklabels=axylabels)
```

As we can see, the entire testing data has been correctly classified, not even a single case of misclassification.

To further evaluate the model, we will use the score() function to obtain the accuracy of the model on the testing set.

```python
# obtaining the accuracy of the model
print(str.format('Accuracy: {:.2f}%', knn.score(x_test, y_test)*100))
```

As we can see, the accuracy of the model is 100%, which is ideal.
