### Q1. Write down the meta information about the “wine.csv” data file available in the UCI repository, https://archive.ics.uci.edu/ml/datasets/Wine.

**Meta-Data** of the Wine Dataset:

These data are the results of a chemical analysis of wines grown in the same region in Italy but derived from three different cultivars. The analysis determined the quantities of 13 constituents found in each of the three types of wines. The attributes are as follows:

- Class: target, the category of cultivator, discrete, values: 1, 2, 3
- Alcohol: the alcohol content in the wine, numerical
- Malic acid: the malic acid content in the wine, numerical
- Ash: the ash content in the wine, numerical
- Alcalinity of ash: the alcanity of ash present in the wine, numerical
- Magnesium: the magnesium content in the wine, numerical
- Total phenols: the total amount of phenols in the wine, numerical
- Flavanoids: the flavanoids content in the wine, numerical
- Nonflavanoid phenols: the content of nonflavanoid phenols in the wine, numerical
- Proanthocyanins: the content of proanthocyanins in the wine, numerical
- Hue: the amount of hue added to the wine, numerical
- OD280/OD315 of diluted wines: a standard measure of the quality of wine, numerical
- Proline: the proline content in the wine, numerical

Total Number of Samples: 178

Samples per class: 59, 71, 48

### Q3. Perform the required preprocessing , 70,30 split and build a classifier using the Multi-Layer Perceptron Classifier model available in SciKit Learn. Use 3 layers with the same number of neurons as there are features along with 500 max iterations.

To solve Question 3, we first check for missing values using the isnull() and sum() functions from the Pandas library.

```python
# checking for missing values
df.isnull().sum()
```

As we can see, there are no missing values. We donot have any categorical features. We donot remove the outliers since the number was insignificant.

After that, we split the data into independent variables (x) and dependent variable (y) and use the train_test_split() function from the sklearn library and divide the dataset into training and testing sets.

```python
# splitting the data into independent and dependent variables
x = df.drop(columns=['Class'])
y = df['Class']

# diving the dataset into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=105)
```

Next, we scale the data. We use the StandardScaler module from the sklearn library to scale the data.

```python
# scaling the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

After that, we build a model using the MLPClassifier module from the sklearn library. To set up 3 layers with the same number of neurons as the number of features, we pass a tuple with the number of columns repeated 3 times to the hidden_layer_sizes parameter of the model. We also set the max_iter parameter to be 500 and fix the random_state to reproduce consistent results. We train this model using the training data.

```python
# building the model
from sklearn.neural_network import MLPClassifier
model = MLPClassifier(hidden_layer_sizes=(x.shape[1], x.shape[1], x.shape[1]), max_iter=500, random_state=105)

# training the model
model.fit(x_train, y_train)
```

### Q4. Tabulate the precision, recall and F1-score.

To solve Question 4, we use the classification_report() function from the sklearn library to get all the metrics - accuracy, precision, recall, f1_score in a tabulare form.

```python
# obtaining the classification report of the model
from sklearn.metrics import classification_report
print(classification_report(y_test, model.predict(x_test)))
```

As we can see, the model has:

- Accuracy: 93%, which means that the model performs well
- Precision: 92%, which means that 92% of the samples predicted to be positive were actually positive
- Recall: 95%, which means that 95% of the total positive samples were predicted to be positive
- F1 Score: 93%, which means that the model is very robust
